```python
import pandas as pd
```

```python
titanic_df = pd.read_csv('data/titanic_clean.csv')
titanic_df.head()
```

```python
titanic_df['fare'].min(), titanic_df['fare'].max()
```

```python
# Boundary values for bins and corresponding labels
bins = [0, 100, 250, 550]
labels = ['low', 'medium', 'high']

titanic_df['fare_category'] = pd.cut(titanic_df['fare'], bins=bins, labels=labels, right=False)
```
