```python
import pandas as pd
```

```python
titanic_df = pd.read_csv('../data/titanic_clean.csv')
titanic_df.info()
```

```python
# Splitting the name column into name & family-name.
split_name = titanic_df['name'].str.split(', ')
titanic_df['family_name'] = split_name.str[0]

# extracting the title from name column
split_title = split_name.str[1].str.split('. ', regex=False)
titanic_df['title'] = split_title.str[0] + '.'
titanic_df['name'] = split_title.str[1]
```

```python
titanic_df.head()
```

```python
# If a passenger was travelling alone or with a family member
titanic_df['alone'] = ((titanic_df['sibsp'] == 0) & (titanic_df['parch'] == 0)).astype(int)
```

```python
# Adding 1 to include the person in the row
titanic_df['family_size'] = titanic_df['sibsp'] + titanic_df['parch'] + 1
```

```python
# deck of the ship extracted from cabin
titanic_df['deck'] = titanic_df['cabin'].str[0]
titanic_df['deck'].replace({'u': 'unknown'}, inplace=True)
```

```python
# number of persons on 1 ticket is not equal to the family size
titanic_df['persons_on_ticket'] = titanic_df.groupby('ticket').transform('count').iloc[:, 0]
titanic_df['fare_per_head'] = titanic_df['fare'] / titanic_df['persons_on_ticket']
```

```python
titanic_df.head()
```

```python
# Persisting extended data to the disk
titanic_df.to_csv('../data/titanic_extended.csv', index=False)
```
