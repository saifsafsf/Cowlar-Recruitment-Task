```python
import pandas as pd
```

```python
titanic_df = pd.read_csv('../data/titanic_clean.csv')
titanic_df.head()
```

```python
# Boundary values for bins and corresponding labels
bins = [0, 21, 50, 100]
labels = ['child', 'adult', 'senior']

titanic_df['age_category'] = pd.cut(titanic_df['age'], bins=bins, labels=labels, right=False)
```
