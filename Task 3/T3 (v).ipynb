```python
import pandas as pd
```

```python
titanic_df = pd.read_csv('data/titanic_extended.csv')
titanic_df.head()
```

```python
# Checking for any abnormal value in the column
titanic_df['sex'].value_counts()
```

```python
sex_dict = {
    'female': 0,
    'male': 1
}

titanic_df['sex'].replace(sex_dict, inplace=True)
```

```python
# Checking for any abnormal value in the column
titanic_df['embarked'].value_counts()
```

```python
embarked_dict = {
    'S': 0,
    'C': 1,
    'Q': 2
}

titanic_df['embarked'].replace(embarked_dict, inplace=True)
```

```python
# Checking for any abnormal value in the column
titanic_df['title'].value_counts()
```

```python
# Abstracting the dictionary-making process
titles = titanic_df['title'].value_counts().index
title_dict = {title:index for index, title in enumerate(titles)}

titanic_df['title'].replace(title_dict, inplace=True)
```

```python
# Checking for any abnormal value in the column
titanic_df['deck'].value_counts().sort_index()
```

```python
decks = titanic_df['deck'].value_counts().sort_index().index
deck_dict = {deck:index for index, deck in enumerate(decks)}

titanic_df['deck'].replace(deck_dict, inplace=True)
```
